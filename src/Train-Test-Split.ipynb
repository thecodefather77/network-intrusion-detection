```python
import pandas as pds
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer
```

# Read the Parquet File

```python
df = pd.read_parquet(r'D:\Analytixlabs\Internship\Project 4\Cyber Security\output\network_data.parquet')
```

# Create the Independent and Target

```python
X = df.drop(columns = ['label','attack','attack_type']).copy()
y_attack = df['attack'].copy()
y_label = df['label'].copy()
y_type = df['attack_type'].copy()
```

# Train Test Split

```python
Xtrain, Xtest, ytrain_attack, ytest_attack, ytrain_label, ytest_label, ytrain_type, ytest_type = train_test_split(
    X,
    y_attack,
    y_label,
    y_type,
    train_size = 0.7,
    test_size = 0.3,
    random_state = 77,
    stratify = y_attack
)
```

# KNN Imputer

```python
imputer = KNNImputer(n_neighbors = 5)
Xtrain_imputed = imputer.fit_transform(Xtrain)
Xtest_imputed = imputer.transform(Xtest)
```

```python
Xtrain_imputed = pd.DataFrame(Xtrain_imputed, columns = Xtrain.columns)
Xtest_imputed = pd.DataFrame(Xtest_imputed, columns = Xtest.columns)
```

```python
ytrain_attack.reset_index(drop = True, inplace = True)
ytest_attack.reset_index(drop = True, inplace = True)
ytrain_label.reset_index(drop = True, inplace = True)
ytest_label.reset_index(drop = True, inplace = True)
ytrain_type.reset_index(drop = True, inplace = True)
ytest_type.reset_index(drop = True, inplace = True)
```

```python
Xtest_imputed = pd.concat([Xtest_imputed, ytest_attack, ytest_type, ytest_label], axis = 1)
```

```python
Xtrain_imputed = pd.concat([Xtrain_imputed, ytrain_attack, ytrain_type, ytrain_label], axis = 1)
```
