```python
# Train Test Split
Xtrain, Xtest, ytrain, ytest = train_test_split(
    X,
    y_attack,
    train_size = 0.7,
    test_size = 0.3,
    random_state = 77,
    stratify = y_attack
)
```

# Baseline Model: Logistic Regression

```python
# Defining the ML Pipeline
pipeline1 = Pipeline(
    steps = [
        ('scaler', StandardScaler()),
        ('model', LogisticRegression(n_jobs = -1))
    ]
)

# Fitting to the Pipeline
pipeline1.fit(Xtrain_imputed, ytrain)

# Predicting the target
ytrain_pred1 = pipeline1.predict(Xtrain_imputed)
ytest_pred1 = pipeline1.predict(Xtest_imputed)

#Train Test Data Performance
print('Train Data Performance:')
print(metrics.classification_report(ytrain, ytrain_pred1))
print('Test Data Performance:')
print(metrics.classification_report(ytest, ytest_pred1))
```

```python
ytest_proba1 = pipeline1.predict_proba(Xtest_imputed)[:,1]
fpr1, tpr1, thresholds1 = metrics.roc_curve(ytest, ytest_proba1)
auc_score1 = metrics.roc_auc_score(ytest, ytest_proba1)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(fpr1, tpr1, label = f"ROC Curve (AUC = {auc_score1:.3f})")
plt.plot([0, 1], [0, 1], linestyle = "--", label = "Random Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROCAUC Curve")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
ytrain_proba1 = pipeline1.predict_proba(Xtrain_imputed)[:,1]
fpr1, tpr1, thresholds1 = metrics.roc_curve(ytrain, ytrain_proba1)
j_scores = tpr1 - fpr1
```

```python
best_threshold = thresholds1[np.argmax(j_scores)]
best_threshold
```

```python
ytest_pred_opt1 = (ytest_proba1 >= best_threshold).astype(int)
print(metrics.classification_report(ytest, ytest_pred_opt1))
```

# Decision Tree

```python
# Defining the ML Pipeline
pipeline4 = Pipeline(
    steps = [
        ('model', DecisionTreeClassifier())
    ]
)

# Fitting to the Pipeline
pipeline4.fit(Xtrain_imputed, ytrain)

# Predicting the target
ytrain_pred4 = pipeline4.predict(Xtrain_imputed)
ytest_pred4 = pipeline4.predict(Xtest_imputed)

#Train Test Data Performance
print('Train Data Performance:')
print(metrics.classification_report(ytrain, ytrain_pred4))
print('Test Data Performance:')
print(metrics.classification_report(ytest, ytest_pred4))
```

```python
# Defining the ML Pipeline
pipeline4 = Pipeline(
    steps = [
        ('model', DecisionTreeClassifier(random_state = 7, class_weight = 'balanced'))
    ]
)

param_grid = {
    'model__criterion': ['gini', 'entropy'],
    'model__max_depth': np.arange(2,16),
    'model__min_samples_split': np.arange(5,51,5),
    'model__min_samples_leaf': np.arange(5,51,5)
}


grid = RandomizedSearchCV(
    estimator = pipeline4,
    param_distributions = param_grid,
    cv = 5,
    scoring = 'roc_auc',
    n_iter = 50,
    random_state = 7,
    n_jobs = -1
)
# Fitting to the Pipeline
grid.fit(Xtrain_imputed, ytrain)

# Best model after tuning
best_model = grid.best_estimator_

# Predicting the target
ytrain_pred4 = best_model.predict(Xtrain_imputed)
ytest_pred4 = best_model.predict(Xtest_imputed)

#Train Test Data Performance
print('Train Data Performance:')
print(metrics.classification_report(ytrain, ytrain_pred4))
print('Test Data Performance:')
print(metrics.classification_report(ytest, ytest_pred4))
```

```python
def objective(trial):

    C = trial.suggest_float("C", 1e-4, 1e2, log=True)
    penalty = trial.suggest_categorical("penalty", ["l1", "l2"])
    solver = trial.suggest_categorical(
        "solver",
        ["liblinear", "saga", "lbfgs"]
    )

    # Reject invalid combinations
    if penalty == "l1" and solver == "lbfgs":
        raise optuna.TrialPruned()

    pipeline = Pipeline(
        steps=[
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(
                C=C,
                penalty=penalty,
                solver=solver,
                class_weight="balanced",
                max_iter=1000,
                n_jobs=-1
            )
        )
    ]
)

    score = cross_val_score(
        pipeline,
        Xtrain_imputed,
        ytrain,
        cv=5,
        scoring="roc_auc",
        n_jobs=-1
    ).mean()

    return score
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)
```
